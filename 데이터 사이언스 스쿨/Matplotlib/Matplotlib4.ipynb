```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sns.set_style("ticks")
sinplot()
```

```python
sns.set_style("darkgrid")
sinplot()
```
