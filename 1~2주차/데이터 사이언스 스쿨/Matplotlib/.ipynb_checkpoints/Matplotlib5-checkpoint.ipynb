```python
iris = sns.load_dataset("iris") 
titanic = sns.load_dataset("titanic")

iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot = 0)
plt.title(iris data bar)
plt.xlabel("Data")
plt.ylabel("Feature")
plt.ylim(0,7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title("Feature")
plt.xlabel("mean")
plt.ylabel("species")
plt.ylim(0, 8)
plt.show()
```

```python
df2.T.plot.bar(rot=0)
plt.title("각 Feature의 종 별 평균")
plt.xlabel("Feature")
plt.ylabel("평균")
plt.show()
```

```python
df3 = titanic.pclass.value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title('ratio')
plt.axis('equal')
plt.show()
```

```python
iris.plot.hist()
plt.title('Features')
plt.xlabel('Data')
plt.show()
```
